```python
import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
import seaborn as sn
import scipy.stats as stats
```

```python
Q7 = pd.read_csv('Q7.csv')
```

```python
median = np.median(Q7["Points"])
median
```

```python
numeric_columns = Q7.select_dtypes(include='number')

# Calculate the median for each numeric column
median_values = numeric_columns.median()

# Display the median values
print("Median values:")
print(median_values)
```

```python
Q7.mode()
```

```python
stats.mode(Q7['Points'])
```

```python
stats.mode(Q7['Score'])
```

```python
stats.mode(Q7['Weigh'])
```

```python
# Convert each column to numeric
numeric_points = pd.to_numeric(Q7['Points'], errors='coerce')
numeric_score = pd.to_numeric(Q7['Score'], errors='coerce')
numeric_weigh = pd.to_numeric(Q7['Weigh'], errors='coerce')

# Combine the numeric columns into a new DataFrame
numeric_df = pd.DataFrame({'Points': numeric_points, 'Score': numeric_score, 'Weigh': numeric_weigh})

# Calculate the variance for each column
variance = numeric_df.var()

# Print or use the variance
print("Variance:\n", variance)
```
