## Recommendation System

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#read the csv file
book=pd.read_csv('book_RS.csv',encoding='ISO-8859-1')
book
```

## EDA

```python
book1=book.iloc[:,1:]
book1
```

```python
#sort by userID
book1.sort_values(['User.ID'])
```

```python
#Renaming the columns
book1=book1.rename({'User.ID':'UserID','Book.Title':'BookTitle','Book.Rating':'BookRating'},axis=1)
book1
```

```python
#number of UserID,Booktitle,Rating
book1.nunique()
```

```python
#number of Ratings ad counts
book1['BookRating'].value_counts().sort_index()
```

```python
#plot the bookrating in barplot
book1['BookRating'].value_counts().sort_index().plot(kind='bar')
```

```python
plt.figure(figsize=(20,8))
#displot
sns.displot(book1.BookRating)
```

```python
#converting long data into wide data using pivote table
book2=book1.pivot_table(index='UserID', columns='BookTitle', values='BookRating').reset_index(drop=True)
book2
```

```python
#Replacing the index values by unique UserId
book2.index=book1['UserID'].unique()
book2
```

```python
#impute those NaNs with 0 values
book2.fillna(0,inplace=True)
book2
```

```python
# Calculating cosine Similarity between Users and array data
user_sim=1-pairwise_distances(book2.values, metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# set the index and column names to UserId
user_sim2.index=book1['UserID'].unique()
user_sim2.columns=book1['UserID'].unique()
user_sim2
```

```python
# Nullifying diagnol values
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
# Most Similar users
user_sim2.idxmax(axis=1)
```

## Consider the most Similar User

```python
def get_recommendation(customer_id):
        
        tem=list(user_sim2.sort_values([customer_id],ascending=False).head(3).index)

        book_list=list(book1[book1['UserID'].isin(tem)]['BookTitle'])

        recommended_books = set(book_list)-set(book1[book1['UserID']==customer_id]['BookTitle'])
        
        return recommended_books
```

```python
get_recommendation(162113)
```
