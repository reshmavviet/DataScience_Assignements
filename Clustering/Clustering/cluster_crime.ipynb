## Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crim1=pd.read_csv('crime_data.csv')
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crim.iloc[:,1:])
```

```python
# Elbow curv
os.environ['OMP_NUM_THREADS'] = '1'
# Ignore the specific warning about KMeans memory leak
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn.cluster")
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, n_init=10)  # Set n_init explicitly
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow curve
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')  # Within-Cluster-Sum-of-Squares
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4, n_init=10 )
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crim1['Clust']=x
```

```python
crim1
```

```python
crim1.iloc[:,1:5].groupby(crim1.Clust).mean()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crim3=pd.read_csv('crime_data.csv')
```

```python
crim3.info()
```

```python
df=crim3.iloc[:,1:5]
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```
